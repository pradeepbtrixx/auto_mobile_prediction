```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Automobile.csv",usecols=['make', 'fuel-type', 'aspiration',
       'num-of-doors', 'body-style','engine-location','height',
       'num-of-cylinders', 'stroke', 'horsepower', 'peak-rpm', 'city-mpg',
       'highway-mpg', 'price'])
df.head(5)
```

```python
#sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df['num-of-doors'].replace('?','two',inplace=True)
```

```python
from numerizer import numerize
for i in df['num-of-cylinders']:
    df['num-of-cylinders'].replace(i,numerize(i),inplace = True) 
```

```python
from numerizer import numerize
for i in df['num-of-doors']:
    df['num-of-doors'].replace(i,numerize(i),inplace = True)
```

```python
df.head(2)
```

```python
df = pd.get_dummies(data = df ,columns=['fuel-type','aspiration','engine-location'],drop_first=True)
```

```python
df.head()
```

```python
for i in df.columns:
    print(i ,len(df[i].unique()))
```

```python
#df.make.value_counts().sort_values(ascending = False).head(10)
```

```python
d = [i for i in df.make.value_counts().sort_values(ascending = False).head(10).index]
print(d)
```

```python
for i in d:
    df['make'+i] = np.where(df['make'] == i,1,0)

df.head()
```

```python
for i in df['body-style'].unique():
    df['body-style'+i] = np.where(df['body-style'] == i,1,0)
df.head()
```

```python
df.drop(['make','body-style'],axis=1,inplace=True)
```

```python
#sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
#sns.pairplot(df,hue=df['price'])
```

```python
df['price'].replace('?',np.nan,inplace=True)
```

```python
df['price'].fillna(13950,inplace=True)
```

```python
df.head()
```

```python
col = ['height','stroke','horsepower','peak-rpm','city-mpg','highway-mpg']
```

```python
for i in col:
    df[i].replace('?',np.nan,inplace=True)
```

```python
for i in col:
    df[i].fillna(df[i].median(),inplace=True)
```

```python
x = df.drop('price',axis=1)
y = df['price']
```

```python
from sklearn.ensemble import RandomForestRegressor
RFR = RandomForestRegressor(random_state=1616)
RFR.fit(x,y)
```

```python
from sklearn.model_selection import cross_val_score
cv = cross_val_score(RFR,x,y,cv = 5,scoring='r2').mean()
cv
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size= 0.3,random_state = 4)
```

```python
RFR.fit(xtrain,ytrain)
pre = RFR.predict(xtest)
```

```python
from sklearn.metrics import r2_score
d = r2_score(pre,ytest)
d
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(xtrain,ytrain)
pr = lr.predict(xtest)
```

```python
r2_score(pr,ytest)
```
